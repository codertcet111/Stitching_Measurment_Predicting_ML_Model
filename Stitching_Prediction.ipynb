```python
import pandas as pd
stitching_file_name = 'stitching_dump.xlsx.csv'
stitching_data = pd.read_csv(stitching_file_name)
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import train_test_split
import sklearn.metrics
knn = KNeighborsRegressor()
regr = MultiOutputRegressor(knn)
```

```python
#Now split the data into test and train data
y = stitching_data[['size_around_ankle','sleeves_length','sleeves_around','front_neck','back_neck','size_around_thigh','size_around_arm_hole','size_around_knee','under_bust']]
X = stitching_data[['shoulder_size','length','chest_size','bottom_length','waist_size','hip_size','weight','height']]
```

```python
# Now Split the data into train and test samples
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.5)
# print(X_test)
# print("this is y test")
# print(y_test)
#Now its time to train our model
regr.fit(X_train, y_train)
```

```python
y_calculated = regr.predict(X_test)
```

```python
# Check for total number of rows predicted
print("Predictade values dimension")
print(y_calculated.shape)
print("**Actual value's dimension")
print(y_test.shape)
# sklearn.metrics.accuracy_score(y_test, y_calculated)
```

```python
# #Now it's time for getting accuracy.
# y_test_duplicate = [row for row in y_test]
# for row in y_test:
#     print(row)
# print(type(y_test))
# y_test.to_numpy()

y_test_values = y_test.values
errors = y_test_values - y_calculated
for index in range(errors.shape[0]):
    print("This is the {0} row ".format(index + 1))
    print("This is actual value: ", end=" ")
    print(y_test_values[index])
    print("This is predicted value: ", end=" ")
    print(y_calculated[index])
    print("This is error : ", end=" ")
    print(errors[index])
    print("",end="\n")
    
```
